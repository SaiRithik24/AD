### Reinforcement Learning: Q-Learning Example

This example demonstrates a simple Q-learning agent. Q-learning is a model-free reinforcement learning algorithm to learn a policy telling an agent what action to take under what circumstances. It learns the value of state-action pairs by iteratively updating a 'Q-table' based on observed rewards and future expected rewards.

```python
import numpy as np
import random

# 1. Define the Environment
# A simple grid-world environment
# S = Start, G = Goal, F = Frozen (negative reward)
# Rewards: G=+1, F=-1, others=0
environment = np.array([
    ['S', ' ', ' ', ' '],
    [' ', 'F', ' ', ' '],
    [' ', ' ', ' ', 'F'],
    [' ', ' ', ' ', 'G']
])

num_states = environment.size
num_actions = 4 # 0: up, 1: down, 2: left, 3: right

# Mapping 2D grid coordinates to a single state index
state_to_index = {}
index_to_state = {}
current_index = 0
for r in range(environment.shape[0]):
    for c in range(environment.shape[1]):
        state_to_index[(r, c)] = current_index
        index_to_state[current_index] = (r, c)
        current_index += 1

# Helper to get the reward for a given state
def get_reward(state):
    r, c = index_to_state[state]
    if environment[r, c] == 'G':
        return 10
    elif environment[r, c] == 'F':
        return -5
    else:
        return -1 # Small negative reward for each step to encourage faster paths

# Helper to determine next state based on action
def take_action(state, action):
    r, c = index_to_state[state]
    new_r, new_c = r, c

    if action == 0: # Up
        new_r = max(0, r - 1)
    elif action == 1: # Down
        new_r = min(environment.shape[0] - 1, r + 1)
    elif action == 2: # Left
        new_c = max(0, c - 1)
    elif action == 3: # Right
        new_c = min(environment.shape[1] - 1, c + 1)

    # If the new state is 'F' or 'G', the episode ends.
    # For simplicity, we just return the new state index and a flag.
    # In a more complex env, reaching 'F' or 'G' would be terminal.
    new_state_index = state_to_index[(new_r, new_c)]
    is_terminal = (environment[new_r, new_c] == 'G' or environment[new_r, new_c] == 'F')

    return new_state_index, is_terminal

print("Environment Grid:")
print(environment)
print("\nState-Index Mapping:")
for k, v in state_to_index.items():
    print(f"  {k} -> {v}")
```

```python
# 2. Initialize Q-table
# Q-table stores Q-values for each state-action pair
# Dimensions: (num_states, num_actions)
Q_table = np.zeros((num_states, num_actions))

# Hyperparameters
learning_rate = 0.8 # Alpha (how much new information overrides old)
discount_factor = 0.95 # Gamma (importance of future rewards)
epsilon = 0.1 # Epsilon (exploration-exploitation trade-off)
num_episodes = 1000 # Number of training episodes

print("Initial Q-table (all zeros):\n", Q_table)
```

```python
# 3. Q-Learning Algorithm

for episode in range(num_episodes):
    current_state = state_to_index[(0, 0)] # Start from 'S'
    done = False

    while not done:
        # Epsilon-greedy strategy: explore or exploit
        if random.uniform(0, 1) < epsilon:
            action = random.randint(0, num_actions - 1) # Explore: choose a random action
        else:
            action = np.argmax(Q_table[current_state, :]) # Exploit: choose action with highest Q-value

        # Take action and observe new state and reward
        new_state, is_terminal = take_action(current_state, action)
        reward = get_reward(new_state)

        # Update Q-value using the Q-learning formula
        # Q(s,a) = Q(s,a) + alpha * (reward + gamma * max(Q(s',a')) - Q(s,a))
        old_q_value = Q_table[current_state, action]
        max_future_q = np.max(Q_table[new_state, :]) if not is_terminal else 0

        new_q_value = old_q_value + learning_rate * (reward + discount_factor * max_future_q - old_q_value)
        Q_table[current_state, action] = new_q_value

        current_state = new_state

        if is_terminal:
            done = True


print("\nTraining complete! Final Q-table:\n")
print(Q_table)

print("\nInterpreting the Q-table: Each row represents a state, and each column represents an action (Up, Down, Left, Right). The values indicate the 'quality' of taking that action from that state, aiming to maximize cumulative reward.")
```

### Test the Learned Policy

Let's see the path the agent takes based on the learned Q-table.

```python
def run_policy(Q_table, start_state_index, max_steps=20):
    current_state = start_state_index
    path = [index_to_state[current_state]]
    total_reward = 0

    print(f"Starting from: {index_to_state[current_state]}")

    for step in range(max_steps):
        # Choose the action with the highest Q-value for the current state
        action = np.argmax(Q_table[current_state, :])

        new_state, is_terminal = take_action(current_state, action)
        reward = get_reward(new_state)
        total_reward += reward

        current_state = new_state
        path.append(index_to_state[current_state])

        print(f"  Step {step + 1}: Took action {action}, moved to {index_to_state[current_state]}, Reward: {reward}")

        if is_terminal:
            print(f"  Reached terminal state {environment[index_to_state[current_state]]}!")
            break
    else:
        print("  Max steps reached.")

    print(f"\nPath taken: {path}")
    print(f"Total Reward: {total_reward}")

start_state_index = state_to_index[(0, 0)] # Start from 'S'
run_policy(Q_table, start_state_index)
```
